```python
import matplotlib.pyplot as plt
import pickle
from sc_scraper import Pilot, Storage, DeltaEntry

%matplotlib inline
plt.rcParams["figure.figsize"] = (14.0, 8.0)
plt.rcParams["figure.dpi"] = 2 * plt.rcParams["figure.dpi"]

data = pickle.load(open("v11.p", "rb"))

color_map = {
    "Mouse": (1, 0, 0),
    "Joystick": (0, 1, 0),
    "Gamepad": (0, 0, 1),
}

def filter_duration(data, threshold):
    new_data = []
    for pilot in data:
        if pilot.flight_time >= threshold:
            new_data.append(pilot)
    return new_data

def filter_matches(data, threshold):
    new_data = []
    for pilot in data:
        if pilot.matches >= threshold:
            new_data.append(pilot)
    return new_data

def filter_device(data, threshold):
    new_data = []
    for pilot in data:
        try:
            fav_input = pilot.favorite_input
            if fav_input[1] >= threshold:
                new_data.append(pilot)
        except KeyError:
            pass
    return new_data

def filter_ship(data, threshold):
    new_data = []
    for pilot in data:
        try:
            fav_input = pilot.favorite_ship
            if fav_input[1] >= threshold:
                new_data.append(pilot)
        except KeyError:
            pass
    return new_data

def ship_names():
    names = {}
    for pilot in data["BR"].get_data().values():
        try:
            names[pilot.favorite_ship[0]] = True
        except KeyError:
            pass
        except TypeError:
            pass
    return sorted(names.keys())

def color_boxplot(bp, color):
    plt.setp(bp["boxes"], color=color)
    plt.setp(bp["caps"], color=color)
    plt.setp(bp["whiskers"], color=color)
    plt.setp(bp["fliers"], color=color)
    plt.setp(bp["medians"], color=color)
```

Ship Usage
==========
The following is a list of all the ships currently available in Arena Commander and their usage in number of players using it as well as the amount of time they are being used.

```python
modes = ["BR", "SB", "VC"]
ships = {}

for mode in modes:
    for pilot in data[mode].get_data().values():
        try:
            tmp = pilot.favorite_ship
            if tmp == None:
                continue
            ship_name = tmp[0]
            if tmp[1] > 0.9:
                if ship_name not in ships:
                    ships[ship_name] = {}
                    for entry in modes:
                        ships[ship_name][entry] = {"count": 0, "time": 0.0}
                ships[ship_name][mode]["count"] += 1
                ships[ship_name][mode]["time"] += pilot.flight_time
        except KeyError:
            pass

#print("Ship                    BR    SB    VC")
#print("--------------------------------------")
#for key in sorted(ships):
#    ship_data = ships[key]
#    print("{:20s} {: 5d} {: 5d} {: 5d}".format(
#        key,
#        ship_data["BR"]["count"],
#        ship_data["SB"]["count"],
#        ship_data["VC"]["count"]
#    ))

for k, mode in enumerate(["BR", "SB", "VC"]):
    sorted_ships = {"count": [], "time": []}
    sorted_keys = sorted(ships.keys())
    for key in sorted_keys:
        ship_data = ships[key][mode]
        sorted_ships["count"].append(ship_data["count"])
        sorted_ships["time"].append(ship_data["time"])
    
    x = range(len(sorted_ships["count"]))
    xt = [v+0.5 for v in x]
    
    plt.figure(2*k + 1)
    plt.title("{} - Ship Usage Count".format(mode))
    plt.bar(x, sorted_ships["count"])
    plt.xticks(xt, sorted_keys, rotation=90)
    plt.ylabel("Number of Users")
    plt.grid()
    
    plt.figure(2*k + 2)
    plt.title("{} - Ship Flight Time".format(mode))
    plt.bar(x, sorted_ships["time"])
    plt.xticks(xt, sorted_keys, rotation=90)
    plt.ylabel("Flight Hours")
    plt.grid()
```

Match Duration
==============

The following plot shows the average match duration distribution.

```python
modes = ["BR", "SB", "VC"]
for i, mode in enumerate(modes):
    y = [e.flight_time / e.matches * 60 for e in data[mode].get_data().values()]
    weights = [e.matches for e in data[mode].get_data().values()]
    plt.subplot(len(modes), 1, i+1)
    plt.xlabel("Match Duration (Minutes)")
    plt.ylabel("Match Counts")
    plt.hist(y, bins=100, weights=weights, range=(0, 60), label=mode)
    plt.legend()
plt.show()
```

Score per Minute Scatter Plot
=============================

The following shows a scatter plot of the players score per minute colored by their preferred input device. Only players that have played more then one hour are considered and those that have a preferred input device with 90% or greated preference.

```python
modes = ["BR", "SB", "VC"]

for i, mode in enumerate(modes):
    plt.figure(i)
    filtered_data = filter_duration(data[mode].get_data().values(), 1.0)
    filtered_data = filter_matches(filtered_data, 4)
    filtered_data = filter_device(filtered_data, 0.9)
    
    #device = [e.favorite_device[0] for e in filtered_data]
    #c = [color_device(e) for e in filtered_data]
    for dev in ["Mouse", "Joystick", "Gamepad"]:
        x = [e.flight_time for e in filtered_data if e.favorite_input[0] == dev]
        y = [e.score_minute for e in filtered_data if e.favorite_input[0] == dev]
        
        plt.scatter(x, y, color=color_map[dev], label=dev)
    plt.title("{} - {:d} Players".format(mode, len(filtered_data)))
    plt.xlabel("Flight Time (hours)")
    plt.ylabel("Score per Minte")
    plt.legend()
    plt.plot()
```

Score per Minute by Ship
========================

Boxplots of the score per minute distribution based on the different ships.

```python
modes = ["BR", "SB"]
devices = ["Mouse", "Joystick", "Gamepad"]
ships = ship_names()

plot_data = {}
for mode in modes:
    plot_data[mode] = {}
    for ship in ships:
        plot_data[mode][ship] = {}
        for dev in devices:
            plot_data[mode][ship][dev] = {"spm": [], "time": [], "kdr": []}

for mode in modes:
    filtered_data = filter_duration(data[mode].get_data().values(), 1.0)
    filtered_data = filter_matches(filtered_data, 4)
    filtered_data = filter_device(filtered_data, 0.9)
    filtered_data = filter_ship(filtered_data, 0.8)
    
    for pilot in filtered_data:
        try:
            ship_name = pilot.favorite_ship[0]
            dev_name = pilot.favorite_input[0]
            
            plot_data[mode][ship_name][dev_name]["time"].append(pilot.flight_time)
            plot_data[mode][ship_name][dev_name]["spm"].append(pilot.score_minute)
            plot_data[mode][ship_name][dev_name]["kdr"].append(pilot.kill_death_ratio)
        except KeyError:
            pass


for k, mode in enumerate(modes):
    sorted_keys = []
    for ship_name, device_data in plot_data[mode].items():
        count = 0
        for entry in device_data.values():
            count += len(entry["time"])
        if count > 250:
            sorted_keys.append(ship_name)
    sorted_keys = sorted(sorted_keys)
    print(sorted_keys)
    
    x = {}
    pos = {}
    for dev in devices:
        x[dev] = []
        pos[dev] = []
    index = 0
    for key in sorted_keys:
        for dev in devices:
            x[dev].append(plot_data[mode][key][dev]["spm"])
            pos[dev].append(index)
            index += 1
    
    print([len(v) for v in x])
    
    plt.figure(2*k + 1)
    for dev in [v for v in x if len(v) > 10]:
        bp = plt.boxplot(x[dev], positions=pos[dev])
        color_boxplot(bp, color_map[dev])
    plt.xlim(-1, index)
    plt.xticks([i*3+1 for i in range(len(sorted_keys))], sorted_keys)
```

Overall Player Stats per Day
============================

Shows statistics over all players that played on a given day.

```python
modes = ["BR", "SB", "VC"]

for i, mode in enumerate(modes):
    date_keys = sorted(data[mode].date_keys())
    
    res = {}
    for key in date_keys:
        res[key] = {
            "flight_time": 0,
            "pilot_count": 0,
            "score": 0,
        }
        for pilot in data[mode].get_data(key).values():
            res[key]["flight_time"] += pilot.flight_time
            res[key]["score"] += pilot.score
            res[key]["pilot_count"] += 1

    plt.subplot(3, 3, 3*i+1)
    plt.title("Flight Time - {}".format(mode))
    plt.bar(range(len(date_keys)), [res[k]["flight_time"] for k in date_keys])
    plt.xticks([i+0.5 for i in range(len(date_keys))], date_keys)
    
    plt.subplot(3, 3, 3*i+2)
    plt.title("Pilot Count - {}".format(mode))
    plt.bar(range(len(date_keys)), [res[k]["pilot_count"] for k in date_keys])
    plt.xticks([i+0.5 for i in range(len(date_keys))], date_keys)
    
    plt.subplot(3, 3, 3*i+3)
    plt.title("Score per Minute - {}".format(mode))
    plt.bar(range(len(date_keys)), [res[k]["score"] / res[k]["flight_time"] for k in date_keys])
    plt.xticks([i+0.5 for i in range(len(date_keys))], date_keys)
    
plt.show()
```

Statistical distribution over Days
==================================

Shows distribution of statistics over players for every day

```python
modes = ["BR", "SB", "VC"]

for i, mode in enumerate(modes):
    date_keys = sorted(data[mode].date_keys())
    
    daily_data = {}
    for key in date_keys:
        daily_data[key] = {
            "flight_time": [],
            "spm": [],
        }
        for pilot in data[mode].get_data(key).values():
            daily_data[key]["flight_time"].append(pilot.flight_time)
            daily_data[key]["spm"].append(pilot.score / pilot.flight_time)

    for key in date_keys:
        plt.figure()
        plt.suptitle("{} - {}".format(mode, key))
        
        plt.subplot(2, 1, 1)
        plt.hist(daily_data[key]["flight_time"], bins=25)
        plt.ylabel("Number of Players")
        plt.xlabel("Flight Time (Hours)")
        
        plt.subplot(2, 1, 2)
        plt.hist(daily_data[key]["spm"], bins=25)
        plt.ylabel("Number of Players")
        plt.xlabel("Score per Minute")
    
        plt.show()
```
