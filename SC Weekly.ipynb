```python
import datetime
import matplotlib.pyplot as plt
import numpy as np
import pickle
from sklearn.neighbors.kde import KernelDensity

from sc_scraper import Pilot, Storage, DeltaEntry

%matplotlib inline
plt.rcParams["figure.figsize"] = (14.0, 8.0)
plt.rcParams["figure.dpi"] = 2 * plt.rcParams["figure.dpi"]

data = pickle.load(open("v11.p", "rb"))

color_map = {
    "Mouse": (1, 0, 0),
    "Joystick": (0, 1, 0),
    "Gamepad": (0, 0, 1),
}
mode_map = {
    "BR": "Battle Royale",
    "SB": "Squadron Battle",
    "VC": "Vanduul Swarm Coop",
}

def filter_duration(data, threshold):
    new_data = []
    for pilot in data:
        if pilot.flight_time >= threshold:
            new_data.append(pilot)
    return new_data

def filter_matches(data, threshold):
    new_data = []
    for pilot in data:
        if pilot.matches >= threshold:
            new_data.append(pilot)
    return new_data

def filter_device(data, threshold):
    new_data = []
    for pilot in data:
        try:
            fav_input = pilot.favorite_input
            if fav_input[1] >= threshold:
                new_data.append(pilot)
        except KeyError:
            pass
    return new_data

def filter_ship(data, threshold):
    new_data = []
    for pilot in data:
        try:
            fav_ship = pilot.favorite_ship
            if fav_ship[1] >= threshold:
                new_data.append(pilot)
        except KeyError:
            pass
    return new_data

def filter_kills(data, threshold):
    new_data = []
    for pilot in data:
        if pilot.kills > threshold:
            new_data.append(pilot)
    return new_data


def filter_remove_ships(data, ship_names):
    new_data = []
    for pilot in data:
        try:
            fav_ship = pilot.favorite_ship
            if fav_ship[0] not in ship_names:
                new_data.append(pilot)
        except KeyError:
            pass
    return new_data


def filter_only_ships(data, ship_names):
    new_data = []
    for pilot in data:
        try:
            fav_ship = pilot.favorite_ship
            if fav_ship[0] in ship_names:
                new_data.append(pilot)
        except KeyError:
            pass
    return new_data

def ship_names():
    names = {}
    for pilot in data["BR"].get_data().values():
        try:
            names[pilot.favorite_ship[0]] = True
        except KeyError:
            pass
        except TypeError:
            pass
    return sorted(names.keys())
```

```python
start_date = datetime.date(2015, 3, 23)
weekly_data = {}

for mode in mode_map:
    week_start = start_date
    cur_date = start_date
    
    weekly_data[mode] = {}
    weekly_data[mode][week_start] = []
    
    while cur_date < datetime.datetime.utcnow().date():
        try:
            cur_date += datetime.timedelta(days=1)
            if (cur_date - week_start) >= datetime.timedelta(days=7):
                week_start = cur_date
                weekly_data[mode][week_start] = []
            
            delta_entries = data[mode].get_data(cur_date)
            weekly_data[mode][week_start].extend(list(delta_entries.values()))
        except KeyError as e:
            pass
```

```python
for m, mode in enumerate(mode_map):
    plot_data = {}
    plot_limits = {}
    
    for key, values in weekly_data[mode].items():
        spm = sorted([max(0.0, entry.score / (entry.flight_time * 60)) for entry in values if entry.flight_time > 0.5])[10:-10]
        flight_time = sorted([max(0.0, entry.flight_time) for entry in values])[10:-10]
        kdr = sorted([entry.kills / entry.deaths for entry in values if entry.deaths > 0 and entry.flight_time > 0.25])[10:-10]

        plot_data[key] = {}
        plot_limits[key] = {}
        
        plot_data[key]["spm"] = [[v,] for v in spm]
        plot_limits[key]["spm"] = (min(spm), max(spm))
        plot_data[key]["flight_time"] = [[v,] for v in flight_time]
        plot_limits[key]["flight_time"] = (min(flight_time), max(flight_time))
        plot_data[key]["kdr"] = [[v,] for v in kdr]
        plot_limits[key]["kdr"] = (min(kdr), max(kdr))
    
    plots = {
        "spm" : ["Score per Minute", 50.0],
        "kdr" : ["Kill Death Ratio", 0.1],
        "flight_time" : ["Flight Time (hours)", 0.1],
    }
    
    for key in sorted(plot_data):
        for i, entry in enumerate(plots):
            bandwidth = plots[entry][1]
            if mode == "VC" and entry == "kdr":
                bandwidth = 1
            
            plt.figure(10*m+i)
            estimator = KernelDensity(kernel="gaussian", bandwidth=bandwidth).fit(plot_data[key][entry])
            y = np.exp(estimator.score_samples(plot_data[key][entry]))
            plt.plot(plot_data[key][entry], y, label="{}".format(key))
            plt.xlim(0.0, max([limit[entry][1] for limit in plot_limits.values()]))
            plt.title("{} - {}".format(mode_map[mode], plots[entry][0]))
            plt.legend()
```
